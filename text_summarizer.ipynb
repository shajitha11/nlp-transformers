```python
from transformers import pipeline
```

```python
summarizer=pipeline("summarization")
```

```python
text = """
Artificial intelligence is a branch of computer science that aims to create machines that can perform tasks that usually require human intelligence.
It includes learning, reasoning, problem-solving, perception, and language understanding.
AI is widely used in industries such as healthcare, finance, automotive, and entertainment.
"""

summary=summarizer(text,max_length=50,min_length=20,do_sample=False)
print(summary[0]['summary_text'])
```
