```python
from transformers import pipeline
```

```python
qa=pipeline("question-answering",model="distilbert-base-uncased-distilled-squad")

context="""
Hugging Face is a company that provides open-source machine learning libraries
and tools. Transformers library is their most popular library, widely used
for NLP tasks like text classification, translation, summarization, and question answering.
"""

question="what is hugging face known for?"
result=qa(question=question,context=context)

print("question:",question)
print("answer:",result['answer'])
print("score:",result['score'])
```
